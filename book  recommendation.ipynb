Stop words

```python
stop = get_stop_words('en')

# remove non-alphanumeric, non-space
stop = [re.sub(r'([^\s\w]|_)+', '', x) for x in stop]
```

```python
def remove_stop(text, stop):
    """
    As function suggests,
    checks fro stop word in the text and doesnt append stop word
    Returns the text without stop words
    """
    custom_stop = stop
    
    #appending a word if it is not a stop word
    new_text = []
    for word in text:
        if word not in custom_stop:
            new_text.append(word)
    return new_text
```

```python
tokenizer = RegexpTokenizer(r'\w+')
p_stemmer = PorterStemmer()
l_stemmer = WordNetLemmatizer()
```

```python
def stem_list(text, p_stemmer):
    """
    Does stemming and lemmataization of the words in the text
    Returns a list of those words
    """
    new_list = []
    for word in text:
        new_list.append(p_stemmer.stem(word))
    return new_list

def lem_list(text, l_stemmer):
    new_list = []
    for word in text:
        new_list.append(l_stemmer.lemmatize(word))
    return new_list
```

```python
def preprocess_text(text):
    
    """
    FLOW-
    1. Remove the unwanted characters
    2. Remove the non alphanueric, non space characters
    4. Remove stop words from the text
    5. Lemmatize the words
    6. Join the words in the list and retrun text
    """
    # remove mixed alphanumeric
    text = re.sub(r"""(?x) # verbose regex
                            \b    # Start of word
                            (?=   # Look ahead to ensure that this word contains...
                             \w*  # (after any number of alphanumeric characters)
                             \d   # ...at least one digit.
                            )     # End of lookahead
                            \w+   # Match the alphanumeric word
                            \s*   # Match any following whitespace""", 
                             "", text)
    
    # remove non-alphanumeric, non-space
    text = re.sub(r'([^\s\w]|_)+', '', text)
    
    # tokenize text
    text = tokenizer.tokenize(text.lower())
    
    # remove stop words
    text = remove_stop(text, stop)
    
    # stem
    text = lem_list(text, l_stemmer)

    
    new_text = ' '.join(text)
    
    
    return new_text
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv=CountVectorizer()
cv_matrix=cv.fit_transform(df.text)
cv_cosine_sim=cosine_similarity(cv_matrix)
```

```python
def get_title_from_index(id):
    """
    Returns the title of the bOOK given the index
    """
    return df[df["Id"]==id]["Book_Title"].values[0]

def get_index_from_title(title):
    """
    Returns the index of the Book given the title
    """
    return df[df.Book_Title == title].Id.values[0]

def get_recommendations(book_id, sim_matrix):
    """
    -Store the recommendations in a list
    -Get the title of book using get_title_from_index
    -Genarate cosine similarity matrix for the book and enumerate it
    -Sort the list based on the similarity
    -Get the index of the book (due to enumeration) 
    and get corresponding title
    -Retrun the list of recs
    """
    recommendations = list()
    
    podcast_title = get_title_from_index(book_id)
    similar_books =  list(enumerate(sim_matrix[book_id]))
    sorted_similar_books = sorted(similar_books,key=lambda x:x[1],reverse=True)
    
    for i in range(4):
        title = get_title_from_index(sorted_similar_books[i][0])
        recommendations.append(title)
    
    return recommendations

def get_dissimilar_recommendations(book_id, sim_matrix):
    """
    -Store the recommendations in a list
    -Get the title of book using get_title_from_index
    -Genarate cosine similarity matrix for the book and enumerate it
    -Sort the list based on the similarity
    -Get the index of the book (due to enumeration) 
    and get corresponding title
    -Retrun the list of recs
    """
    recommendations = list()
    
    podcast_title = get_title_from_index(book_id)
    similar_books =  list(enumerate(sim_matrix[book_id]))
    sorted_similar_books = sorted(similar_books,key=lambda x:x[1])
    
    for i in range(4):
        title = get_title_from_index(sorted_similar_books[i][0])
        recommendations.append(title)
    
    return recommendations
```
